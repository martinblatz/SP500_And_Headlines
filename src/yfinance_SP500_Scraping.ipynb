```python
import yfinance as yf
```

```python
# create a ticker object using the S&P index identifier
idx = yf.Ticker("^GSPC")

# retrieve stock info. There is another "slow" method, but it doesn't appear 
# to be operational at the time
idx.fast_info
```

```python
# get historical market data
# Valid period values:
# 1d, 5d, 1m, 3m, 6m, 1y, 5y, 10y, ytd, max

# This only works backwards from the day the data is retrieved
# use 10 years to obtain the 5 years we are interested in (2018-2022) 
# with some unnecessary data on either end
hist = idx.history(period="10y")
hist
```

```python
# Use the colab capability to save the df to the VM, 
# then download it to my local machine so I don't have to repeat this process 

from google.colab import files
hist.to_csv('SP500 Hist.csv', encoding = 'utf-8-sig') 
files.download('SP500 Hist.csv')
```
